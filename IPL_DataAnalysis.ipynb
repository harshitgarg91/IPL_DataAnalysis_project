```python
# Getting the frequeny of result column
ipl['result'].value_counts()
```

```python
# finding the toss winner with resprct to each team
ipl['toss_winner'].value_counts()
```

```python
# Extracting the records where a winning team bat first
batting_first=ipl[ipl['win_by_runs']!=0]
```

```python
batting_first.head()
```

```python
batting_first.tail()
```

```python
plt.figure(figsize=(5,5))
plt.hist(batting_first['win_by_runs'])
plt.title("Distribution of Runs")
plt.xlabel('Runs')
plt.show()
```

```python
# finding out the number of wins w-.r.t each team after batting first
batting_first['winner'].value_counts()
```

```python
# making a bar plot for 3 teams with most wins after batting first
plt.figure(figsize=(6,4))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),batting_first['winner'].value_counts()[0:3],width=.35,color=['yellow','blue','purple'])
plt.grid()
plt.show()
```

```python
# making pie chart for distribution of most wins after bat first
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
# extracting those records where a team has won after batting second
bat_sec = ipl[ipl['win_by_wickets']!=0]
bat_sec
```

```python
# making a histogram for frequency of wins w.r.t no of wickets
plt.figure(figsize=(5,5))
plt.hist(bat_sec['win_by_wickets'],bins=30)
plt.show()
```

```python
# finding out the frequency of number of wins w.r.t each time afte batting second
bat_sec['winner'].value_counts()
```

```python
# making a bar plot for 3 teams with most wins after batting second
plt.figure(figsize=(6,4))
plt.bar(list(bat_sec['winner'].value_counts()[0:3].keys()),bat_sec['winner'].value_counts()[0:3],width=.35,color=['red','purple','blue'])
plt.grid()
plt.show()
```

```python
# making pie chart for distribution of most wins after bat second
plt.figure(figsize=(7,7))
plt.pie(list(bat_sec['winner'].value_counts()),labels=list(bat_sec['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
